```python
# concant and appennd df (from Sentdex):

import pandas as pd

df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])

# df1 and df2 have the same columns - it added to the bottum:
concat = pd.concat([df1,df2])
print(concat)

concat = pd.concat([df1,df2,df3])
print(concat)

# append - add to the end:
df4 = df1.append(df2)
print(df4)

df4 = df1.append(df3)
print(df4)

# add a series:
s = pd.Series([80,2,50], index=['HPI','Int_rate','US_GDP_Thousands'])
df4 = df1.append(s, ignore_index=True)
print(df4)
```

```python
# join (with index) and merge:
import pandas as pd

df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Unemployment':[7, 8, 9, 6],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])

# df1 and df3 share the culomn 'HPI' - df3 culomns were added
print(pd.merge(df1,df3, on='HPI'))

# df1 and df2 share 'HPI' and 'Int_rate' so we can merge them with both culomns:
# Notice here there are two versions of the US_GDP_Thousands. This is because we didn't share on these columns!
print(pd.merge(df1,df2, on=['HPI','Int_rate']))

# merge df1 and df3 to a new df and set 'HPI' as index:
df4 = pd.merge(df1,df3, on='HPI')
df4.set_index('HPI', inplace=True)
print(df4)
```

```python
# merge and join when the values of ech df are differnt:
# 2 df with the same 'Year' culomn but the values are not the same:
df1 = pd.DataFrame({
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55],
                    'Year':[2001, 2002, 2003, 2004]
                    })

df3 = pd.DataFrame({
                    'Unemployment':[7, 8, 9, 6],
                    'Low_tier_HPI':[50, 52, 50, 53],
                    'Year':[2001, 2003, 2004, 2005]})

# merge them on 'Year'culomn - not all values are in merged df:
merged = pd.merge(df1,df3, on='Year')
print(merged)

#merge them on 'Year' column as an index - still data is lost!:
merged = pd.merge(df1,df3, on='Year')
merged.set_index('Year', inplace=True)
print(merged)

#correct merge of dfs whithout losing Data:
# Merging on the left is literally on the left dataframe. 
#We had df1, df3, the one on the left is the first one, df1. 
#So, we wound up with an index that was identical to the left dataframe (df1).
merged = pd.merge(df1,df3, on='Year', how='left')
merged.set_index('Year', inplace=True)
print(merged)

# Merging on right - so this time the index here is from the right (df3).
merged = pd.merge(df1,df3, on='Year', how='right')
merged.set_index('Year', inplace=True)
print(merged)

#merging on outer - the index 'Year' is from both df1 and df3!
merged = pd.merge(df1,df3, on='Year', how='outer')
merged.set_index('Year', inplace=True)
print(merged)

#merging with inner - merge by the shared indexes! this is the defoult of merge:
merged = pd.merge(df1,df3, on='Year', how='inner')
merged.set_index('Year', inplace=True)
print(merged)
```

# exersice 5 - Obesity Tredns in the UK

```python
import quandl
import pandas as pd
api_key = open('quandlapikey.txt','r').read()

df_OV = quandl.get("OECD/HEALTH_LVNG_BODYVBMS_TOTPOPTX_GBR", authtoken=api_key)
df_Cal = quandl.get("OECD/HEALTH_LVNG_FOODCALT_CALPERQT_GBR", authtoken=api_key)
print(df_OV.head())
print(df_Cal.head())
```

```python
# In each df, change the column name "Value" to "% Obese or Overweight" or "Kcal Per Capita Per Day"
df_OV=df_OV.rename(columns = {'Value':'% Obese or Overweight'})
df_Cal=df_Cal.rename(columns = {'Value':'Kcal Per Capita Per Day'})

#print(df_OV.head())
#print(df_Cal.head())
df_OV.to_csv('Obeisty')
df_Cal.to_csv('Calories')
```

```python
# Use df1.join(df2) to join these two dfs by their index
DataF= df_Cal.join(df_OV, on='Date', how='left')
DataF
```

```python
# Count how many null values exist in the joined table
count_null= DataF.isnull().values.sum()
count_null
```

```python
# 3.
#  What is the standard deviation of the mean calorie intake in the UK?
# numpy.std(a, axis=None, dtype=None, out=None, ddof=0, keepdims=<no value>, *, where=<no value>)
# Pandas uses the unbiased estimator (N-1 in the denominator), whereas Numpy by default does not. 
# To make them behave the same, pass ddof=1 to numpy.std().
intake_cal_std = np.std(DataF, ddof=1)
intake_cal_std
```

```python
# What is the correlation coefficient between obesity and caloire intake? Pandas
# for entire df
Obis_Cal_Corr= DataF.corr()
print(Obis_Cal_Corr)
```

```python
# Separate Series instead:
corr= DataF['% Obese or Overweight'].corr(DataF['Kcal Per Capita Per Day'])
print(corr)
```

```python
# With SciPy module
import scipy.stats
# drop null from DataF
DataF = DataF.dropna()

x = DataF['% Obese or Overweight'] 
y = DataF['Kcal Per Capita Per Day'] 
scipy.stats.pearsonr(x, y) 
```

```python
plt.scatter(x,y)
plt.show()
```

```python
slope, intercept, r, p, stderr = scipy.stats.linregress(x, y)
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
line
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='s', label='Data points')
ax.plot(x, intercept + slope * x, label=line)
ax.set_xlabel('% Obese or Overweight')
ax.set_ylabel('Kcal Per Capita Per Day')
ax.legend(facecolor='white', loc=4)
plt.show()
```
