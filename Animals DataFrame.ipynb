```python
# 1. Import numpy as np and pandas as pd and show pandas version:


import numpy as np
import pandas as pd
print(pd.show_versions())

```

```python
# few of the fundamental routines for selecting, sorting, adding and aggregating data in DataFrames.

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# 1.Create a DataFrame from this dictionary data which has the index labels

df_animals = pd.DataFrame(data, index=labels)
df_animals
```

```python
# 2.Display a summary of the basic information about this DataFrame and its data
df_animals.describe()
```

```python
# 3.Return the first 3 rows of the DataFrame df
df_animals.head(3) # equivalently:  df.iloc[:3]
#df_animals.head()
```

```python
# 4. Select just the 'animal' and 'age' columns.
print(df_animals[['animal', 'age']]) # or:  df.loc[:, ['animal', 'age']]
```

```python
# 5. Select the data in rows [3, 4, 8] and in columns ['animal', 'age']
# Access group of rows and columns by integer position(s)

data_select=df_animals.iloc[[3,4,8], [0, 1]] # df.ix[[3, 4, 8], ['animal', 'age']]
print(data_select)
```

```python
# 6. Select the rows where the number of visits is greater than 3
rows_greater_3 = df_animals[df_animals['visits']>3]
print(rows_greater_3)
```

```python
# 7. Select the rows where the age is missing, i.e. is NaN
rows_no_age= df_animals[df_animals['age'].isnull()]
print(rows_no_age)
```

```python
# 8. Select the rows where the animal is a cat and the age is less than 3
cat_smaller_3 = df_animals[(df_animals['animal'] == 'cat') & (df_animals['age']<3)]
print(cat_smaller_3)
```

```python
# 9. Select the rows the age is between 2 and 4 (inclusive)
# Series.between(left, right, inclusive='both')
#Return boolean Series equivalent to left <= series <= right.

# NA values are treated as False.
# This function is equivalent to (left <= ser) & (ser <= right)

# With inclusive set to "neither" boundary values are excluded:

#this line give a bool:
#middle_age = df_animals['age'].between(2,4)
#print(middle_age)

middle_age= df_animals[df_animals['age']. between(2,4)]
print(middle_age)
```

```python
# Sort of columns:

# 6. Sort df first by the values in the 'age' in decending order, then by the value in the 'visit' column in ascending order.
# Hint: You can add ascending=[False, True] to sort_values()
df_animals.sort_values(by=['age'], ascending=[False])

```

```python
# 7. The 'priority' column contains the values 'yes' and 'no'. 
# Replace this column with a column of boolean values: 'yes' should be True and 'no' should be False.
# map
# map accepts a dict or a Series. 
# Values that are not found in the dict are converted to NaN, unless the dict has a default value (e.g. defaultdict):


df_animals['priority'] = df_animals['priority'].map({ 'yes': 'True', 'no': 'False' })
df_animals
```

```python
# 8. In the 'animal' column, change the 'snake' entries to 'python'.
df_animals['animal'] = df_animals['animal'].replace('snake', 'python')
df_animals
```
